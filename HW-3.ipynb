```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_1 = np.array([10,10,10])
```

```python
def norm(x):
    norm = 0
    for i in x:
        norm += i**2
    norm = norm**(1/2)
    return norm 
```

```python
norm(v_1)
```

```python
x = np.linspace(-10, 10, 100)
y = 1/x

plt.plot(x,y)
```

```python
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
X, Y = np.meshgrid(x,y)
F = X**2 + Y**2 - 0.6
plt.contour(X,Y,F,[0])
plt.show()
```

```python
z = -np.linspace(9,15,100)
x = np.linspace(-26,26,1000)

x,z = np.meshgrid(x,z)

Z = -np.exp(-0.05*z) +4*(z+10)**2 
X = x**2


plt.contour(x,z,(X+Z),[0])
plt.xlim([-1.5,1.5])
plt.ylim([-11.5,-8.5])
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)

Z = 2*X + 5*Y

z = 2*X + 5*Y + 20
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,z)
show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X,Y = np.meshgrid(X,Y)

Z = X**2 + Y**2

ax.plot_wireframe(X,Y,Z)
show()
```

## Графики на плоскость

```python
x = np.linspace(-2*np.pi,3*np.pi,201)
k = 6
a = 2
b = 5
y = k*np.cos(x-a) + b
```

```python
plt.plot(x,y)
plt.grid(True)
plt.title('Для параметров k = 6,a = 2, b = 5')
```
